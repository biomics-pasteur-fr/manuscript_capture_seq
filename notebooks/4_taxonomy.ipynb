# Utility functions

```python
%load_ext autoreload
%autoreload 2

import glob
import json
import pandas as pd
from sequana import MultiKrakenResults2, MultiKrakenResults
from pylab import axhline, savefig, linspace
```

```python
def get_files_and_samples(case, PATH="."):
    samples = [4885,4716,4660,4520,4707,4697,4676,4653,4861,4787,4688,4673,4777,4668,4489,4798,4797,4656,4544]
    samples = [str(x) for x in samples][::-1]
    filenames = [f"{PATH}/taxonomy/{case}/{sample}/summary.json" for sample in samples]
    return filenames, samples


def get_files_and_samples_meta(PATH="."):
    samples = [4885,4716,4660,4520,4707,4697,4676,4653,4861,4787,4688,4673,4777,4510,4668,4489,4798,4797,4656,4544, "Cneg_Water"]
    samples = [str(x) for x in samples][::-1]
    json_files = [f"{PATH}/taxonomy/meta/{sample}/summary.json" for sample in samples]
    csv_files = [f"{PATH}/taxonomy/meta/{sample}/kraken.csv" for sample in samples]
    return json_files, csv_files, samples

```

```python
from  matplotlib import cm
L= 6
cmap="viridis"
colors6 = [cm.get_cmap(cmap)(x) for x in linspace(0.2, 1, L)][::-1]
colors5 = [colors6[0]] + colors6[2:]
colors4 = [colors6[0]] + colors6[3:]

```

# Illumina resp v2

```python
filenames, samples = get_files_and_samples("illumina_resp_v2")      
k = MultiKrakenResults2(filenames, sample_names=samples)
df = k.plot_stacked_hist("test.png", dpi=200, ytick_fontsize=12, cmap='viridis_r', sorting_method=None, fontsize=16, colors=colors6)
for x in [14.5,10.5,6.5,3.5]:
    axhline([x], lw=4, color="k")
savefig("taxonomy_Ill_resv_v2.png", dpi=200)
savefig("taxonomy_Ill_resv_v2.pdf", dpi=200)
savefig("taxonomy_Ill_resv_v2.eps", dpi=200)
```
